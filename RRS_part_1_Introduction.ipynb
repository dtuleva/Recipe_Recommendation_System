## 1.1. Introduction

The task of recommendation systems is to provide personalized suggestions to users, helping them discover relevant items from a vast pool of options. These systems leverage data analysis and machine learning to predict user preferences and optimize the user experience and therefore boost site traffic and/or sales. There are many different types of recommendation systems - collaborative (recommends items based on user behavior patterns), content-based (based on item's features and similarity), hybrid (combines different recommendation approaches), session-based (based on co-occurrences of items or interaction order in sessions), demographic-based and other.

Common methods used in recommender systems are factor models and neighborhood methods. Factor models work by decomposing the sparse user-item interactions matrix to a set of d dimensional vectors one for each item and user in the dataset. [[1]](#l1)

Here is an example of matrix factorization, houses being the items:

[image source](https://medium.com/@rebirth4vali/implementing-matrix-factorization-technique-for-recommender-systems-from-scratch-7828c9166d3c)

Deep neural networks can be used to add rich feature representations of both users and items, leading to more diverse and sophisticated recommenders. Here is an visualization of the popular two-tower model architecture - each tower being a embedding model for a feature or an item. Both towers could be an item embedding, when preforming item-to-item recommendations. In our model we follow this approach and have a user tower - user id and timestamp of when a review was made and an item tower - recipe id, name and image.

[image source](https://medium.com/@pauloyc/tensorflow-recommenders-for-powerful-recommendation-system-e3dec138a07f)

Adding rich features to the model helps to mitigate the cold start problem - when we need predictions for unknown user, new item or when there is insufficient data to build the system. Another aproach to deal with cold start is to combine different types of information - clicks from sessions (that can be from  unregistered / unknown users), ratings, reviews, item similarity and other.
