This paper introduces adaptive optimization method with greatly reduced memory overhead, allowing parameter adaptivity for large models and batch sizes.

Adaptively tuning the learning rate for each parameter during the optimization process using cumulative second-order statistics (as Adam or Adagrad optimizers) is the norm in all state-of-the-art models. However the overhead of the optimizer becomes quite substantial for larger models, effectively reducing the possible batch size. To counteract this problem, the authors suggest novel optimizer algotithm - Square-root of Minima of Sums of Maxima of Squared-gradients Method, or in short, SM3. This is the pseudocode:

Generally the algorithm makes a collection of sets as a cover of the parameters, each parameter index may be contained in multiple sets, each set is assigned single variable. This way the algorithm has number of variables proportional to the number of sets and not the number of original parameters(weghts). In large scale applications this allowes substantial savings in memory. For each set in the cover, the algorithm maintains a running sum of the maximal variance over all gradient entries. For each parameter the minimum over all variables is associated with all the sets it's index is in. The learning rate corresponding to each gradient is determined by taking the square-root of this minimum.
